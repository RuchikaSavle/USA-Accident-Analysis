```python
data=data.dropna(subset=['Precipitation(in)'])    
```

```python
data=data.dropna(subset=['Temperature(F)','Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Direction', 'Wind_Speed(mph)',
                      'Weather_Condition'])
```

```python
data=data.dropna(subset=['City','Sunrise_Sunset',
       'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight'])
```

```python
data_cat=data.select_dtypes('object')
data_num=data.select_dtypes(np.number)
data_cat=data_cat.drop('ID',axis=1)
```

```python
data_cat=data.select_dtypes('object')
col_name=[]
length=[]

for i in data_cat.columns:
    col_name.append(i)
    length.append(len(data_cat[i].unique()))
data_2=pd.DataFrame(zip(col_name,length),columns=['feature','count_of_unique_values'])
data_2
```

```python
data.drop(['Description','Zipcode','Weather_Timestamp'],axis=1,inplace=True)
```

```python
accidents_by_cities = data['City'].value_counts()
accidents_by_cities
```

```python
accidents_by_cities[:10]

```

```python
fig, ax = plt.subplots(figsize=(8,5))
accidents_by_cities[:10].plot(kind='bar')
ax.set(title = 'Top 10 cities By Number of Accidents',
       xlabel = 'Cities',
       ylabel = 'Accidents Count')
plt.show()
```

```python
data['Weather_Condition'].value_counts()
```

```python
accidents_severity = data.groupby('Severity').count()['ID']
accidents_severity
```

```python
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
label = [1,2,3,4]
plt.pie(accidents_severity, labels=label,
        autopct='%1.1f%%', pctdistance=0.85)
circle = plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
ax.set_title("Accident by Severity",fontdict={'fontsize': 16})
plt.tight_layout()
plt.show()
```
